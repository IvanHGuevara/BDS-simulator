```python
# I want to read the cases from the file, determine the grid's edges, 
# and then plot the cases by binning them together.


```

```python
contents = pandas.read_csv('../processed_cases.csv')
```

```python
min_la = min(contents['latitude'])
```

```python
max_la = max(contents['latitude'])
```

```python
min_lo=min(contents['longitude'])
```

```python
max_lo=max(contents['longitude'])
```

```python
import numpy as np
```

```python
lats  = np.linspace(min_la, max_la, 20)
```

```python
longs = np.linspace(min_lo, max_lo, 20)
```

```python
lats
```

```python
longs
```

```python
xs = []

for long in longs:
    for _ in range(len(lats)):
#         print(long)
        xs.append(long)
```

```python
ys = [] 

for _ in range(len(longs)):
    for lat in lats:
#         print(lat)
        ys.append(lat)
```

```python
# These are created because the width of the bar can outpass its edges for some reason. 

width = longs[1] - longs[0]
height = lats[1] - lats[0]
```

```python
# I want to set the colors to green for fewer cases, red for more cases, and white for no cases.

colors = ['#%06d' %(x%(255*255*255)) for x in range(0, 50*len(lats)*len(longs), 50)]
colors = ['#0099ff']*len(lats)*len(longs)
hm = figure()
hm.rect(xs, ys, width=width, height=height, line_color='white', fill_color=colors, )
show(hm)
```
